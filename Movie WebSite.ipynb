```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = "https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now/2/"
```

```python
response = requests.get(base_site)
response
```

```python
html = response.text
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
divs = soup.find_all("div", {"class":"col-sm-18 col-full-xs countdown-item-content"})
```

```python
len(divs)
```

```python
divs[1].find("a").text
```

```python
movie_name = [i.find("a").text for i in divs]
```

```python
year = [i.find("span",{"class":"subtle start-year"}).text.strip("()") for i in divs]
```

```python
score = [i.find("span",{"class":"tMeterScore"}).text for i in divs]
```

```python
# class="descriptor"
directors = [i.find("div",{"class":"director"}) for i in divs]
directors[0]
```

```python
[i.find_all("a") for i in directors]
```

```python
director = []

for i in directors:
    director_links = i.find_all("a")
    director_name = [link.text for link in director_links]
    result = " , ".join(director_name)
    
    director.append(result)
```

```python
director
```

```python
director = [" , ".join([link.text for link in i.find_all("a")]) for i in directors]
```

```python
director
```

```python
import pandas as pd
```

```python
movie_info = pd.DataFrame()

movie_info["Movie Title"] = movie_name
movie_info["Year"] = year
movie_info["Score"] = score
movie_info["Director"] = director

movie_info
```

```python
movie_info.to_excel("movie data.xlsx", index = False)
```
